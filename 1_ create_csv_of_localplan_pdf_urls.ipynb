```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://local-plans-prototype.herokuapp.com/local-plans"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
results = soup.find_all(class_="lps-table-view__heading")
urls = list()

for result in results:
    urls.append(
        "https://local-plans-prototype.herokuapp.com" + result.get("href"))

pdfs = list()
for url in urls:
    page = requests.get(url)

    soup = BeautifulSoup(page.content, "html.parser")

    try:
        pdf = soup.find_all(
            class_="data-source")[0].find_all("a")[0].get("href")
        lpa = soup.find_all(class_="govuk-heading-xl")[0].getText()

    except:
        print(url)
        print(soup.find_all(class_="data-source"))

    pdfs.append({'pdf': pdf, 'lpa': lpa})

print(pdfs)
```

```python
import pandas
df = pandas.DataFrame.from_dict(pdfs)
df.to_csv("./localplans.csv", sep=',', index=False)
```
